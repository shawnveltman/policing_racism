```python
import glob
import os
import pandas as pd
```

```python
class ReportManager:
    def __init__(self):
        pass

    def consolidate_reports(self, directory,extension='.csv'):
        files = glob.glob(directory + "/*" + extension)
        master_report = pd.DataFrame()
        for file in files:
            summary = pd.read_csv(file,index_col='county_fips')
            master_report = pd.concat([master_report, summary])

        master_filename = directory + "/master_report.csv"
        master_report.to_csv(master_filename)
```

```python
directory = "data/summaries"
filepath = directory + "/master_report.csv"

if os.path.exists(filepath):
    os.remove(filepath)

reporter = ReportManager()
reporter.consolidate_reports(directory)

master_report = pd.read_csv(filepath)
```
